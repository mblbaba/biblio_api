```python
import requests
def add(b_id):
    
    bu = f"http://127.0.0.1:5000/books/{b_id}/copy/add"
    r=requests.post(bu, {"book_id": b_id}, headers={"Content-Type": "application/json"})
    print(r.json()["message"])
    
    
k = 399-201
for i in range(201, 399):
    for _ in range(10):
        try:
            add(i)
        except Exception as e:
            print("une erreur s'est produite", e)
    k -= 1
    print("restant : ", k)
    
```
